```python
X = dataset.drop(columns='Sales',axis=1)
Y = dataset[['Sales']]
print(X)
print(Y)

xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.3,random_state=0)
print("Xtrain shape :",xtrain.shape)
print("Xtest shape :",xtest.shape)
```

```python
model = LinearRegression()
model.fit(xtrain,ytrain)
print("Model :",model)

prediction = model.predict(xtest)
print("Prediction :",prediction)
```

```python
pred = prediction.ravel()
y_test = ytest.values
plt.figure(figsize=(10, 6))
sns.regplot(x=pred, y=y_test, ci=None, line_kws={"color": "red"})
plt.xlabel('Predicted values')
plt.ylabel('Actual values')
plt.title('Regression Best Fit Line on Predictions and Test Values')
plt.show()
```

```python
score = r2_score(ytest,prediction)*100
mae   = mean_absolute_error(ytest,prediction)
mse   = mean_squared_error(ytest,prediction)

print(f"R2 score            : {score:.2f}")
print(f"Mean Absolute Error : {mae:.2f}")
print(f"Mean Sqaured Error  : {mse:.2f}")
```
